## Visualize model Architecture -2

```python
import torch
from torchviz import make_dot
from torch.autograd import Variable
import torch.nn as nn

# Assuming you have a GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create a dummy input for a grayscale image of size 28x28
x = Variable(torch.randn(1, 1, 28, 28)).to(device)

# Move the model to the device
model = TinyCNN(params.dropout_rate).to(device)

# Pass the input through the model
y = model(x)

# Visualize the model
make_dot(y).render("cnn_arch", format="png")

from IPython.display import Image
Image(filename='cnn_arch.png') 
```
